````python
#1) Seu código aqui

def get_dados(UF: str,
              tipo: str):
    
    url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={UF}&tipo_interessado={tipo}'
    r = requests.get(url)
    data_json = r.json()
    df = pd.DataFrame(data_json['items'])
    df
    return df

df = get_dados(UF = 'MG', tipo = 'Estado')
df.head(5)
````

````python
# 2) Seu código aqui

df_MG = get_dados(UF = 'MG', tipo = 'Estado')
df_MG.loc[df_MG['status'] == 'Arquivado por decurso de prazo']['status'] .value_counts()
````

````python
# 3) Seu código aqui

df_BH = get_dados(UF = 'BA', tipo = 'Estado')
df_BH.loc[df_BH['status'] == 'Deferido']['status'] .value_counts()
````
